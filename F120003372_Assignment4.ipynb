```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
req=requests.get(' https://www.sfu.ca/computing/people/faculty.html')
```

```python
bsoup=BeautifulSoup(req.text,'lxml')
```

After analysing the structure and CSS of the webpage, I noticed that the names and designations of the staff where in "h4" tag, so I extracted them accordingly.

```python
bsoup.prettify()
names=list()
des=list()
for i in bsoup.find_all('h4'):
    x=(''.join(i.findAll(text=True)))
    p=x.find(',')
    names.append(x[:p])
    m=x[p+1:]
    if(m.find('\n') != -1):
        m=m.replace('\n','')
    des.append(m)

```

I organised it into a Data frame, thereby finishing Task 1.

```python
import pandas as pd
profiles=pd.DataFrame({
    "Faculty Name" : names,
    "Designation" : des
})
profiles
```

Task 2 calls for visiting a staff's profile page and then retrieving their area of interest in terms of research. The code below shows the extraction of staff profile links. Once I got my hands on the links, I created another requests object that contained that link and from there another BeautifulSoup object.

```python
import re
areas=list()
for link in bsoup.findAll('a', attrs={'href': re.compile("http://")}):
    #print(link.get('href'))
    samp=link.get('href')
    if(samp.find('.html')!=-1):
        areas.append(samp)

for prof in areas:
    if(prof.find('/computing/people/faculty/')==-1):
        areas.remove(prof)

req1=requests.get(areas[1])
bsoup1=BeautifulSoup(req1.text,'lxml')
```

The code below extracts the staff's area of interest.

```python
res=list()
for i in bsoup1.select("div.ruled"):
        for ix in i.select("ul li",text=True):
            x=(''.join(ix.findAll(text=True)))
            x=x.replace('\n','')
            res.append(x)
            
            
```

And I organised the staff's interests in a data frame.

```python
l=list()
for ind in res[0:3]:
    l.append(ind)
    l.append(' , ')
stri=''.join(l)
pd.set_option('max_colwidth', 100)
research=pd.DataFrame({
    "Faculty Name " : names[4],
    "Research Interests " : stri
},index=[0])
research
```
